```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/World_population'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

print(soup.title)
print(soup.title.text)

table = soup.find_all('table')

dataframe = [] # empty list 

for i, table in enumerate(table):
    rows = table.find_all("tr")[1:]
    data = []
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele]) # Get rid of empty values
    dataframe.append(data)

df = pd.DataFrame(dataframe[3])
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/World_population'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# print(soup.title)
# print(soup.title.text)

table = soup.find_all('table')

dataframe = [] # empty list 

for table_element in table:
    rows = table_element.find_all("tr")[1:]
    data = []
    for row in rows:
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        data.append(cols) # Get rid of empty values
    df = pd.DataFrame(data)
    dataframe.append(df)

dataframe[1].to_csv('webscrap.csv', index=False,header=True)
# print the first few lines to see what
```
