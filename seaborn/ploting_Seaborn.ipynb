# seaborn and ploting 

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### important thing for plotting according to grammar of graphics 

1. Data (df)
2. Type of plot 
3. X axis 
4. Y axis 
5. Grouping (hue/color) based on categorical variable 
6. Size of the plot 
7. Color palletes of the plot
8. Styel of the plot 
9. How do you interpreter 

```python
# data  
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
sns.scatterplot(data = df ,x='age' ,y='fare', hue='sex')
```

```python
# scatter plot
# scatter plot size of figure size 

plt.figure(figsize=(8,4))
sns.scatterplot(data = df ,x='age' ,y='fare', hue='sex')
```

```python
# scatter plot
# scatter plot size of figure size 

plt.figure(figsize=(16,6))


# change the scatter points colors
palette = {'male': 'green', 'female': 'pink'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', size='fare',palette=palette)
# labels 
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('scatter plot of Age vs Fare')

# save a plot 300dpi , 600dpi , 150pdi
# plt.savefig('./outputs/scatter_plot.png', dpi=300)

# # saving a plot in vactor file 
# plt.savefig('./outputs/scatter_plot.svg')
```

```python
# styel of the plot 
sns.set_theme(style='darkgrid',font_scale=1.2,font='Times New Roman')
palette = {'male': 'green', 'female': 'pink'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex',palette=palette)
# labels 
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('scatter plot of Age vs Fare')
```

Interpertation: 
1. we have Age of pessenger at x-axis and fare at y-axis
2. The scatter plot show the relation between age and fare of titanic pessenger.
3. we further grouped the pessenger by their gender such as male (green points) and 
female (pink points)
4. Two male pessenger gave the highest fare in titanic, and they might be travelling first class.
5. The olderst person at titanic is also a male having 80 years of age  as show in scatter plot. (As show in figure.1 )

## **box plot**

```python
# styel of the plot 
plt.figure(figsize=(12,4))
sns.set_theme(style='darkgrid',font_scale=1.2,font='Times New Roman')
palette = {'First': 'green', 'Second': 'blue','Third':'red'}
sns.boxplot(data=df, x='sex', y='age', hue='class',palette=palette)
# labels 
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title(' plot box of Age vs Fare')

# change legend position and title
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
# plt.legend(loc="lower right", title='class Name')
```

```python
plt.figure(figsize=(12,4))
sns.set_theme(style='darkgrid',font_scale=1.2,font='Times New Roman')
palette = {'First': 'green', 'Second': 'blue','Third':'red'}
sns.boxenplot(data=df, x='sex', y='age', hue='class',palette=palette)
# labels 
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title(' plot box of Age vs Fare')

# change legend position and title
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
# plt.legend(loc="lower right", title='class Name')
```
