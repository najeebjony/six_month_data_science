# Cross Validation 

## Introduction
Cross-validation is a powerful technique used in machine learning to assess the predictive performance of a model. It involves splitting the input data into two
sets, training the model on the training set, and then evaluating its performance on the test set. The test set is used to assess the model's performance

## Why Cross Validation?
Cross-validation is primarily used in applied machine learning to estimate the skill of a machine learning model on unseen data. That is, to use a limited sample in order to estimate how the model is expected to perform in general when used to make predictions on data not used during the training of the model.

# K-Fold Cross Validation:
K-Fold Cross Validation is a common type of cross validation that is widely used in machine learning. It helps to solve the problem of overfitting in machine learning models. It is a resampling technique used to evaluate machine learning models on a limited data sample. It is a procedure used to estimate the skill of the model on new data. It is called K-Fold because the original sample is randomly partitioned into K equal sized subsamples. Of the K subsamples, a single subsample is retained as the validation data for testing the model, and the remaining K-1 subsamples are used as training data. The cross-validation process is then repeated K times, with each of the K subsamples used exactly once as the validation data. The K results from the folds can then be averaged to produce a single estimation. The advantage of this method is that all observations are used for both training and validation, and each observation is used for validation exactly once.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

# load the iris dataset
iris = load_iris()

nb = GaussianNB()

# perform 10-fold cross validation with k=5
scores = cross_val_score(nb, iris.data, iris.target, cv=5, scoring='accuracy')

print("scores for each fold", scores)
print("mean score", scores.mean())
print("standard deviation", scores.std())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn .model_selection import train_test_split
import pandas as pd
import numpy as np 
import seaborn as sns
# load dataset
tips =sns.load_dataset('tips')

# split into train and test
X = tips[['total_bill','tip','size']]
y = tips['sex']

# Define the Gaussian Naive Bayes model
model = GaussianNB()

# perform k-fold cross validation with k=5
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')

# print the scores for each fold and the mean score
print("Scores for each fold:", scores)
print("Mean score:", scores.mean())
print("Standard Deviation:", scores.std())
```

# How To plot k-fold CV 

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# load the iris dataset
iris = load_iris()
X = iris.data
y= iris.target

clf = DecisionTreeClassifier()
#perform k-fold cross-validation with k=8

scores = cross_val_score(clf,X,y,cv=8)

#plot the result 
plt.figure(figsize=(8,6))
plt.plot(range(1,9), scores, color='blue',linestyle='dashed',marker='o',)
plt.title('K-Fold Cross-Validation Result')
plt.xlabel('Fold Number')
plt.ylabel('accuracy')
plt.ylim(0.5,1.0)
plt.show()

```
