```python
import numpy as np 

np.random.seed(42)
m = 100
X = 2 * np.random.rand (m,1)
y = 4 + 3 *  X  + np.random.randn(m,1)


# Normal Equation
theta_best = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(theta_best)

# make a scatter plot 
import matplotlib.pyplot as plt
plt.plot(X,y, 'b.')
plt.plot(X, X.dot(theta_best), 'r-')
plt.show()
```
