```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
sns.scatterplot(data=df,x = 'age', y='fare', hue='sex')
```

```python
# scatter plot 
# scatter plot size of figure size 

plt.figure(figsize=(8,4))
sns.scatterplot(data=df,x = 'age', y='fare', hue='sex')
```

```python
# scatter plot 
# scatter plot size of figure size 
plt.figure(figsize=(8,4))

#change the scatter points colors
palette = {'male':'green','female':'red'}
sns.scatterplot(data=df,x = 'age', y='fare', hue='sex',palette=palette)
# labels 
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot Of Age vs Fare')

# saving a plot 300dpi , 600dpi , 150dpi

plt.savefig('../outputs/scatter_plot.png',dpi=300)

# saving a plot in vactor file such as svg
plt.savefig('../outputs/scatter_plot.svg')
```

sns set styel search from seaboern web/library 

```python
# styel of a plot 
sns.set_theme(style= 'darkgrid',font_scale=1.5,font='Times New Roman')
plt.figure(figsize=(16,6))

#change the scatter points colors
palette = {'male':'green','female':'red'}
sns.scatterplot(data=df,x = 'age', y='fare', hue='sex',size='fare',palette=palette)
# labels 
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot Of Age vs Fare')
```

Interpertion:
1. we have age of pessenger of x-axis and fare at y-axis 
2. The scatter plot show the realationship between age and fare of titanic pessenger.
3. we further grouped the pessenger by their genders such as male (Green points) and female (Red points)
4. Two male pessengers gave the highest fare in titanic , and the might be travelling first class
5. The olderst person at titanic is also a male having 80 years of age as show in the scatter plot.(AS show in figure. 1 )



# **Box Plot**

```python
sns.set_theme(style= 'darkgrid',font_scale=1,font='Times New Roman')
plt.figure(figsize=(9,4))

#change the scatter points colors
palette = {'First':'green','Second':'red','Third':'blue'}
sns.boxplot(data=df,x = 'sex', y='age', hue='class',palette=palette)
# labels 
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('box plot Of Age vs Fare')

# change legend position and title
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.legend(loc='upper left',title='class Name')


```

```python
sns.set_theme(style= 'darkgrid',font_scale=1,font='Times New Roman')
plt.figure(figsize=(9,4))

#change the scatter points colors
palette = {'First':'green','Second':'red','Third':'blue'}
sns.boxenplot(data=df,x = 'sex', y='age', hue='class',palette=palette)
# labels 
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('boxen plot Of Age, based on Gender and class Name of Pessenger')

# change legend position and title
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
# plt.legend(loc='upper left',title='class Name')

# change the x and y tick lables
plt.xticks(ticks=[0,1],labels=["Male","Female"], rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# make scatter plot with plotly
import plotly.express as px

fig = px.scatter(df, x="age", y="fare", color='class')
fig.show()
```

```python
# bar plot use seaborn
sns.set_theme(style= 'darkgrid',font_scale=1,font='Times New Roman')
plt.figure(figsize=(12,4))

#change the scatter points colors
palette = {'First':'green','Second':'red','Third':'blue'}
sns.barplot(data=df,x = 'sex', y='age', hue='class',palette=palette,capsize=0.1)
# labels 
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('bar plot Of Age, based on Gender and class Name of Pessenger')

# change legend position and title
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.legend(loc='upper right',title='class Name')

# change the x and y tick lables
plt.xticks(ticks=[0,1],labels=["Male","Female"], rotation=45)
plt.yticks(rotation=45)

plt.show()
```

```python
sns.set_theme(style= 'darkgrid',font_scale=1,font='Times New Roman')
plt.figure(figsize=(12,4))

#change the scatter points colors
palette = {'First':'green','Second':'red','Third':'blue'}
sns.violinplot(data=df,x = 'sex', y='age', hue='class',palette=palette)
# labels 
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('violinplot plot Of Age, based on Gender and class Name of Pessenger')

# change legend position and title
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.legend(loc='upper right',title='class Name')

# change the x and y tick lables
plt.xticks(ticks=[0,1],labels=["Male","Female"], rotation=45)
plt.yticks(rotation=45)

plt.show()
```
