# **OpenCv and Images**

```python
# import libraries 

import numpy as np 
import cv2 as cv
```

```python
# read the image 

img = cv.imread("./images/test.png")
```

```python
from matplotlib import pyplot as plt
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# check the size of image
img.shape
```

```python
# convert the img grayscale and then check the shape 

gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
gray_img.shape

```

```python
#read img 
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2GRAY))
plt.axis('off')
plt.show()

```

```python
# change the size and read 

resize_img = cv.resize(img, (500, 500))
plt.imshow(cv.cvtColor(resize_img, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# read webcam using cv
cap = cv.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    cv.imshow("Webcam", frame)
    
    if cv.waitKey(1) == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
